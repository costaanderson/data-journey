```python
df = pd.read_csv('data/players_data_light-2024_2025.csv',index_col=0, parse_dates=True)
df
```

Manipulação de dados(EDA)

```python
df['Nation'] = df['Nation'].apply(lambda x: str(x).split()[-1])
df
```

```python
#Quantidade de jogadores por liga
print("\nJogadores por liga")
print(df['Comp'].value_counts())

#Quantidade de jogadores por posição
print("\nJogadores por posição")
print(df['Pos'].value_counts())
```

```python
#Ajustando a principal posição dos jogadores
#Utilizando o método split para separar as posições por ',' e buscar a primeira posição do atleta, utilizando o apply para aplicar a função lambda em toda a coluna 
df['Pos'] = df['Pos'].apply(lambda x: x.split(',')[0])

print("\nJogadores por posição")
print(df['Pos'].value_counts())
```

```python
#Informações básicas para seleção de atletas entre 20 e 26 anos com minutagem acima de 900 minutos e não seja goleiro
df_filtred = df[
    (df['Age'] < 24) &
    (df['Min']>= 900) &
    (df['Pos']!= 'GK')
]
df_filtred 
```
